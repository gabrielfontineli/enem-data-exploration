## Análise das notas por região do Brasil

```python
import pandas as pd
df = pd.read_parquet("../data/interim/enem_2023.parquet")

# Exemplo 1: média por região
df.groupby("REGIAO_NOME_PROVA")["NOTA_MEDIA_5"].mean().sort_values(ascending=False)

# Exemplo 2: boxplot de notas por região (matplotlib, sem seaborn)
import matplotlib.pyplot as plt
df.boxplot(column="NOTA_MEDIA_5", by="REGIAO_NOME_PROVA", rot=45)
plt.title("Distribuição da NOTA_MEDIA_5 por Região (local da prova)")
plt.suptitle("")
plt.xlabel("Região")
plt.ylabel("Nota média (5 áreas)")
plt.show()

# Exemplo 3: tabelão rápido por UF
(df.groupby(["SG_UF_PROVA"])
   ["NOTA_MEDIA_5"].agg(["count","mean","std","median"])
   .sort_values("mean", ascending=False))
```
